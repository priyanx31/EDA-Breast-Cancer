```python
import pandas as pd
```

```python
cancer=pd.read_csv("C:\\Users\\Lenovo PC\\Desktop\\python\\dataset\\breast_cancer.csv")
cancer.head()
```

```python
#get dummies=is used to convert strings to numbers and cerates various columns
```

```python
x=pd.get_dummies(cancer['class'])
x.head()
```

```python
#dropna will detete null values and drop will delete anything
```

```python
#drop class column
cancer.drop(['class'],axis=1,inplace=True)
```

```python
cancer.head()
```

```python
#now fill the dummy variables in the dataset
cancer=pd.concat([cancer,x],axis=1)
cancer.head()
```

```python
#label encoder(=alphabetical order which consists of first alphabet=0,2nd alphabet=1,3rd=2,same vlaue=same index)#modification of get dummies because various columns are not created rather changed in that column
#label encoder is a package of sklearn(scientific kit learn)
```
